# Matplotlib

Matplotlib es una libreria de python que permite realizar graficas 2D apartir de datos de un array o una lista.

**Documentacion de Matplotlib:** https://matplotlib.org/


**Grafico Lineal**

sirve para graficar tendencias de valores a lo largo del tiempo (meses, a√±os, dias, periodos)

````
# importamos la libreria de matplolib
import matplotlib.pyplot as plt 

mes = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']  
temperatura = [25, 29, 18, 22, 12, 30]

# plot me permite graficar, los parametros que recibe este metodo son dos listas: (eje x, eje y) 
plt.plot(mes, temperatura)

# esta funcion me permite mostrar la grafica
plt.show()
````

**Personalizacion de graficas lineales**

````
# importamos la libreria de matplolib
import matplotlib.pyplot as plt

mes = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']  
temperatura = [25, 29, 18, 22, 12, 30]

#aqui agregamos mas datos a las dos listas existentes 
mes = mes + ['Jul', 'Ago', 'Sep'] 
temperatura = temperatura + [21, 11, 27]

plt.plot(mes, temperatura)

# creamos las dos variables que almacenen el nombre de cada eje
xlab = "Mes"
ylab = "Temperatura"

# creamos la variable que almacena el titulo de la grafica
title = "Temperaturas de los meses en el a√±o 2020" 

# con la funcion xlabel y ylabel podemos colocar las etiquetas a cada eje
plt.xlabel(xlab) 
plt.ylabel(ylab)

# con la funcion title podemos colocar el titulo
plt.title(title)

# esta funcion permite definir valores de intervalos
# en los ejes x o y, en este caso estan siendo aplicado para el eje y
plt.yticks([10, 15, 20, 25, 30]) 
                                
plt.grid(True)
plt.show()
````

**Graficos de dispersion**
<p>ideales para realizar evaluaciones de si existe una correlacion entre dos variables</p>

````
# nos permite graficar un grafico de dispersion
# en la funcion scatter podemos utilizar varios argumentos por ejemplo s = tama√±o de los puntos , c = color 
plt.scatter(mes, temperatura, s = 50, c = 'purple')

plt.show()
````

**Histogramas**
<P> permiten visualizar la distribucion de frecuencia de una variable.

````
# aqui graficamos el histograma, (valores, numerosDeBins) si no especificamos 
# el numero de bins matplotlib los pondra por defecto que es 10
plt.hist(temperatura)

plt.show()

# limpiamos el histograma para volver a graficarlo
plt.clf()

# aqui el bins lo especificamos y queremos que sea de 5
plt.hist(temperatura, 5)

plt.show()
````

**Pandas**
<br>
* Es una herramienta de manipulacion de datos de alto nivel, construida sobre la libreria de Numpy.
*   Los Dataframe son las estruturas mas importantes de pandas, estas nos permiten guardar datos tabulares donde se les puede asiganar etiquetas a las columnas y filas.

````
#listas
pelicula = ['Par√°sitos', 'Joker', 'El irlandes', 'Jo jo Rabbit','Mujercitas', 'Historia de un matrimonio', '1917']
premios = [4,2,0,1,1,1,3]
ganador_del_oscar =  [True, False, False, False, False, False, False]

#importamos las libreria de pandas
import pandas as pd

#aqui creamos un diccionario apartir de las listas anteriormente mencionadas
my_dict = {'pelicula': pelicula,
           'premios': premios,
'ganador_del_oscar': ganador_del_oscar
}

# construimos un dataframe apartir de el diccionario creado anteriormente
cine = pd.DataFrame(my_dict)

#imprimimos el nuevo Dataframe
print(cine)

#ahora queremos cambiarle las etiquetas a las filas y lo haremos de la siguiente forma
print("******************************************")
print("nuevo Dataframe con etiquetas en las filas")
print("******************************************")

#definimos esta lista
row_labels = ['PAR', 'JOK', 'EIR', 'JOJ', 'MUJ', 'HIS', 'MIL']

#luego le asigamos en los indices del diccionario cine la anterior lista
cine.index = row_labels
print(cine)

````

**Operadores Booleanos en Numpy**

````
import numpy as np
my_house = np.array([18.0, 20.0, 10.75, 9.50])
your_house = np.array([14.0, 24.0, 14.25, 9.0])

#ahora vamos a realizar unas comparaciones con los operadores logicos, pero aplicados a las anteriores matrices de numpy

#queremos saber Qu√© √°reas de my_house son mayores 18.5 o menores que 10, lo hacemos asi:
print(np.logical_or(my_house > 18.5, my_house < 10))

#queremos saber Qu√© √°reas son m√°s peque√±as que 11 en ambos my_house y your_house, lo hacemos asi:
print(np.logical_and(my_house < 11, your_house < 11))
````

**Bucles sobre Matriz Numpy**
<p>Si estamos tratando con una matriz 1D Numpy, recorrer todos los elementos es mas simple lo podemos hacer asi</p>


```
for x in my_array :
    ...
```

<p>Si se trata de una matriz Numpy 2D, es m√°s complicado. Una matriz 2D est√° formada por varias matrices 1D. Para iterar expl√≠citamente sobre todos los elementos separados de una matriz multidimensional, lo haremos asi:</p>


```
for x in np.nditer(my_array) :
    ...
```





**Bucles sobre Dataframe**
<p>Para iterar sobre un dataframe lo debemos hacer con la funcion iterrows(), usando el ciclo for, cada observaci√≥n se repite una y otra vez, la etiqueta de la fila y el contenido real de la fila est√°n disponibles</p>

```
for etiqueta, fila in dataframe.iterrows() :
    ...
```



````
import pandas as pd
cine = pd.read_csv('cine.csv', index_col = 0)

# aqui iteramos sobre las filas del dataframe cars
for etiqueta, fila in cine.iterrows():
    print(etiqueta)
    print(fila)
````

````
# aqui tenemos otra opcion de como poder iterar sobre filas especificas
for etiqueta, fila in cine.iterrows() :
    print(etiqueta + ": " + str(fila['premios']))
````

**Agregar columnnas**
<p>si queremos agregar una columna en nuestro dataframe a partir de algun calculo de las columnas originales lo podemos hacer de la siguiente forma</p>

````
# aqui usamos un ciclo for  para agregar una nueva columna, nombrada DIRECTOR, 
# que contenga una versi√≥n en may√∫sculas de los nombres de los directores de cada pelicula
# en la columna "director". usamos la funcion .upper() para aplicar lo anteriormente mencionado
for lab, row in cine.iterrows():
    cine.loc[lab, "DIRECTOR"] = row["director"].upper()
print(cine)
````

````
#aqui tenemos otra forma mas eficiente, para obtener los mismo resultados, ya que iterrows()
# es un poco ineficiente en un Dataframe de pandas
#(str.upper) realiza lo mismo que .upper(), pero la primera se aplica
# cuando queremos realizar el procedimiento de esta manera
cine["DIRECTOR"] = cine["director"].apply(str.upper)
print(cine)
````
